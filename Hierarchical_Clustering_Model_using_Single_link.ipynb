```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

# Data points
data = np.array([[1, 10], [7, 11], [5, 6], [9, 7], [2, 5], [3, 9], [2, 12], [8, 8], [8, 10], [2, 4], [3, 6], [3, 3]])

# Hierarchical clustering with single linkage
linked = linkage(data, method='single', metric='euclidean')

# Plot dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linked, labels=[f'A{i+1}' for i in range(len(data))], orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram (Single Linkage)')
plt.show()
```
